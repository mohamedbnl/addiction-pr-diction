# Prediction du niveau d'adiction au réseau sociaux 

### Source: https://www.kaggle.com/datasets/muhammadroshaanriaz/time-wasters-on-social-media/data
### Ce dataset est particulièrement utile pour analyser les corrélations entre les habitudes d'utilisation des réseaux sociaux et des facteurs tels que l'âge, le revenu, la profession ou encore le niveau de dépendance. 

### ***Déscription des caractéristiques:***


- UserID : Identifiant unique attribué à chaque utilisateur.

- Age : Âge de l’utilisateur.

- Gender : Genre de l’utilisateur.

- Location : Localisation géographique de l’utilisateur.

- Income : Revenu annuel de l’utilisateur.

- Debt : Indique si l’utilisateur a des dettes (oui ou non).

- Owns Property : Indique si l’utilisateur possède un bien immobilier.

- Profession : Profession ou métier de l’utilisateur.

- Demographics : Catégorie démographique générale de l’utilisateur (Rural, Urban).

- Platform : Réseau social principalement utilisé.

- Total Time Spent : Temps total passé sur les réseaux sociaux.

- Number of Sessions : Nombre de fois que l’utilisateur se connecte à la plateforme.

- Frequency : Fréquence d’utilisation.

- Scroll Rate : Vitesse de défilement du contenu pendant l’utilisation.

- Productivity Loss : Degré de perte de productivité causée par l’utilisation des réseaux sociaux.

- Satisfaction : Niveau de satisfaction de l’utilisateur vis-à-vis de l’expérience sur la plateforme.

- Self Control : Niveau de maîtrise de soi de l’utilisateur face à l’usage des réseaux sociaux.

- Addiction Level : Degré de dépendance aux réseaux sociaux.

- Current Activity : Activité en cours de l’utilisateur lorsqu’il utilise les réseaux sociaux.

- Watch Reason : Raison principale pour laquelle l’utilisateur regarde du contenu.

- Video ID : Identifiant unique d’une vidéo regardée.

- Video Category : Catégorie de la vidéo.

- Video Length : Durée de la vidéo en minutes.

- Time Spent On Video : Temps passé à regarder une vidéo spécifique.

- Number of Videos Watched : Nombre total de vidéos regardées par l’utilisateur.

- Engagement : Niveau d’interaction avec le contenu.

- Importance Score : Note indiquant à quel point l’utilisateur juge le contenu utile ou pertinent.

- Watch Time : Moment de la journée où l’utilisateur regarde des vidéos.

- Device Type : Type d’appareil utilisé.

- OS : Système d’exploitation de l’appareil.

- Connection Type : Type de connexion Internet utilisée (ex. : Wi-Fi, données mobiles).

## Oversampling des données train

```python
from collections import Counter
print("Training data class distribution:", Counter(y_train))
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# Original training data distribution
print("Original training data distribution:", Counter(y_train))
# Counter({2: 161, 5: 158, 0: 120, 3: 101, 1: 41, 7: 38, 4: 26, 6: 23})

# STEP 1: Oversample the actual minority classes (4, 6, 7, 1)
target_minority = 100  # Target for minority classes
sampling_strategy_over = {1: target_minority, 4: target_minority, 
                          6: target_minority, 7: target_minority}

smote = SMOTE(sampling_strategy=sampling_strategy_over, random_state=42, k_neighbors=5)
X_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)
print("After oversampling minorities:", Counter(y_train_resampled))

# STEP 2: Undersample the actual majority classes (2, 5, 0, 3)
target_majority = 100  # Target for majority classes
sampling_strategy_under = {0: target_majority, 2: target_majority, 
                           3: target_majority, 5: target_majority}

rus = RandomUnderSampler(sampling_strategy=sampling_strategy_under, random_state=42)
X_train_balanced, y_train_balanced = rus.fit_resample(X_train_resampled, y_train_resampled)
print("After balancing:", Counter(y_train_balanced))

# Visualize the balanced training data distribution
plt.figure(figsize=(10, 6))
classes = sorted(Counter(y_train_balanced).keys())
counts = [Counter(y_train_balanced)[c] for c in classes]
plt.bar(classes, counts)
plt.title('Training Data Class Distribution after Custom Resampling')
plt.xlabel('Addiction Level')
plt.ylabel('Count')
plt.grid(True, alpha=0.3)
plt.show()
```

## Tester diférents algorithmes d'entrainement

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(multi_class='multinomial', solver='lbfgs', class_weight='balanced')
model.fit(X_train_balanced, y_train_balanced)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

precision= precision_score(y_test, y_pred, average='macro')
print("Precision:", precision)
```

### ===> L'accuracy du model avec la Regression Logistic est de 80% 

```python
from sklearn.model_selection import cross_val_score, KFold, cross_validate, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
import numpy as np

# Define a pipeline that includes your resampling steps and model
# This ensures resampling is applied per fold
pipeline = ImbPipeline([
    # First, oversample the minority classes
    ('oversample', SMOTE(sampling_strategy={1: 80, 4: 80, 6: 80, 7: 80}, 
                         random_state=42, k_neighbors=5)),
    # Then, undersample the majority classes
    ('undersample', RandomUnderSampler(sampling_strategy={0: 100, 2: 100, 3: 100, 5: 100}, 
                                      random_state=42)),
    # Finally, train the model
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
])

# Use StratifiedKFold instead of KFold to maintain class distribution in each fold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation (using multiple metrics)
cv_results = cross_validate(
    pipeline, x, y, 
    cv=skf,
    scoring=['accuracy', 'f1_macro', 'precision_macro', 'recall_macro'],
    return_train_score=True
)

# Print all scores
print("Cross-validation scores:")
for metric in ['test_accuracy', 'test_f1_macro', 
               'test_precision_macro', 'test_recall_macro']:
    scores = cv_results[metric]
    print(f"{metric}: {scores}")
    print(f"Mean {metric}: {scores.mean():.4f}")
    print(f"Std {metric}: {scores.std():.4f}")
    print("")
```

### ===> On remarque d'apres le matrice de confusion qu'il ya un peut d'erreur au niveau de quelque classes

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier

# Binarize the output (for ROC)
y_test_bin = label_binarize(y_test, classes=[0,1,2,3,4,5,6,7])
n_classes = y_test_bin.shape[1]

# Get probability scores for all classes
y_score = model.predict_proba(x_test)

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot all ROC curves
plt.figure(figsize=(8, 6))
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f'Classe {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Courbes ROC Multiclasse')
plt.legend()
plt.show()
```
